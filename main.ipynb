```python
# Импортирование необходимой библиотеки для работы с данными
import pandas as pd
```

```python
# Чтение csv файлов с данными о пользователях, создание DataFrames
df1 = pd.read_csv('Customer Acqusition.csv')
df2 = pd.read_csv('Repayment.csv')
df3 = pd.read_csv('spend.csv')
```

```python
# Очистка данных от дубликатов
df1.drop_duplicates(keep='last', inplace=True)
df2.drop_duplicates(keep='last', inplace=True)
df3.drop_duplicates(keep='last', inplace=True)
```

```python
# Удаление колонки из DataFrame, которая содержала только значения Nan
df2.drop(columns='Unnamed: 4', inplace=True)
```

```python
# Очистка данных 
df1.dropna(inplace=True)
df2.dropna(axis=0, inplace=True)
df3.dropna(inplace=True)
```

```python
# Группировка данных по "Сustomer" и переименование колонки с итоговыми результатами
repayment = df2.groupby('Customer')['Amount'].sum().reset_index()
repayment.rename(columns={'Amount': 'Total_repayment'}, inplace=True)
repayment
```

```python
# Группировка данных по "Сustomer" и переименование колонки с итоговыми результатами
spend = df3.groupby('Customer')['Amount'].sum().reset_index()
spend.rename(columns={'Amount': 'Total_spend'}, inplace=True)
```

```python
# Объединения данных с помощью Merge по колонке "Сustomer"
result = df1.merge(spend)
result = result.merge(repayment)
```

```python
# Сохранение результата в формате csv файла
result.to_csv('output.csv', index=False, sep=',')
```
